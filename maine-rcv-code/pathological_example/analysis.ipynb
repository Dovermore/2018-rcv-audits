# For 4 candidates

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
file_name = "audit_actual_004050000249902500.csv"
data = pd.read_csv(file_name)
del data["Unnamed: 0"]
```

```python
winner = "C"
data["upset_probability"] = 1 - data.loc[:, winner]
```

```python
# Box plot as described: https://en.wikipedia.org/wiki/Box_plot 
sns.set(font_scale=1.5) 
plt.figure(figsize=(20, 15))
sns.set_style('whitegrid')
ax = sns.boxplot(x='sample_size',y='upset_probability', data=data,fliersize=1)
ax.set_xlabel("Upset Probability",fontsize=20)
ax.set_ylabel("Sample Size",fontsize=20)
```

### Check two extremes

```python
data[data.sample_size==100].describe()
```

```python
data[data.sample_size==3000].describe()
```
