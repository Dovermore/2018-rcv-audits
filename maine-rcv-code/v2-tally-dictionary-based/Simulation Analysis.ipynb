## Visual Distribution of Win Probability Across Seed

```python
ax = df_100[df_100.columns[:-2]].plot(title="Win Probability Per Candidate",figsize=(20, 10),marker='.',lw=0, markersize=10,fontsize=20)
ax.title.set_size(25)
ax.set_xlabel("Simulation / Seed",fontsize=20)
ax.set_ylabel("Win Probability",fontsize=20)
```
